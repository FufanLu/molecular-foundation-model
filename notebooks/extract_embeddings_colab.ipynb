```python
import sys
sys.path.append('molecular-foundation-model')

from src.dataset.load_leffingwell import load_leffingwell
from src.preprocessing.clean_smiles import clean_smiles

df = load_leffingwell()
df = clean_smiles(df)
print(f"Loaded {len(df)} molecules")
print(df.head())
```

```python
from unimol_tools import UniMolRepr

clf = UniMolRepr(data_type='molecule', remove_hs=False)
smiles_list = df['smiles'].tolist()

print(f"Extracting embeddings for {len(smiles_list)} molecules...")
reprs = clf.get_repr(smiles_list)

print(f"CLS embedding shape: {reprs['cls_repr'].shape}")  # expected: (N, 512)
```

```python
import pickle
import numpy as np

embeddings = dict(zip(df['compound'].tolist(), reprs['cls_repr']))

with open('leffingwell_embeddings.pkl', 'wb') as f:
    pickle.dump(embeddings, f)

print(f"Saved {len(embeddings)} embeddings to leffingwell_embeddings.pkl")
```
