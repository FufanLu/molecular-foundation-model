```python
import sys
sys.path.append('molecular-foundation-model')

from src.evaluation.similarity import load_embeddings
from src.evaluation.neighbors import neighbor_report
from src.evaluation.label_consistency import label_consistency, global_consistency
from src.evaluation.visualization import plot_umap
from src.dataset.load_leffingwell import load_leffingwell
from src.preprocessing.clean_smiles import clean_smiles

embeddings = load_embeddings()
df = load_leffingwell()
df = clean_smiles(df)
print(f"Loaded {len(embeddings)} embeddings, {len(df)} molecules")
```
